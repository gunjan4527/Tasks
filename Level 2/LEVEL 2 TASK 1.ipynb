```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load the dataset
dataset = pd.read_csv(r"C:\Users\gunjan kumar\OneDrive\Documents\Dataset .csv")
dataset.head()
```

ANALYZE THE DISTRIBUTION OF AGGREGATE RATINGS AND DETERMINE THE MOST COMMON RATING RANGE

```python
# Find aggregate ratings

aggregate_ratings = dataset['Aggregate rating']
rating_counts = aggregate_ratings.value_counts().sort_index()
```

```python
# Create histogram

plt.hist(dataset['Aggregate rating'], bins=20, edgecolor='black')
plt.title("The Distribution of Aggregate Ratings")
plt.xlabel("Aggregate rating")
plt.ylabel("Frequency")
plt.show()
```

```python
# Find the most common rating range

most_common_rating_range = rating_counts.idxmax()
print("The most common rating range is", most_common_rating_range)
```

THE AVERAGE NUMBER OF VOTES RECEIVED BY RESTAURANTS

```python
# find the average number of votes received by restaurants

average_votes = dataset['Votes'].mean()
print("The average number of votes received by restaurentsis", average_votes)
```
