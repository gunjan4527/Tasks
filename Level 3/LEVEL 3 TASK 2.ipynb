```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load the dataset
dataset = pd.read_csv(r"C:\Users\gunjan kumar\OneDrive\Documents\Dataset .csv")
dataset.head()
```

```python
# Identify the restaurant with the highest and lowest number of votes

highest_votes = dataset.loc[dataset["Votes"].idxmax()]
lowest_votes = dataset.loc[dataset["Votes"].idxmin()]
```

```python
# Display their datails

highest_votes[["Restaurant Name", "Votes", "Aggregate rating"]], lowest_votes[["Restaurant Name", "Votes", "Aggregate rating"]]
```

```python
# Check corrlation between Votes and Aggregate rating

correlation = dataset["Votes"].corr(dataset["Aggregate rating"])
```

```python
# Visualize the relationship using a scatter plot

plt.figure(figsize=(10,5))
sns.scatterplot(x=dataset["Votes"], y=dataset["Aggregate rating"], alpha=0.5)
plt.xlabel("Number of votes")
plt.ylabel("Aggregate rating")
plt.title("Correlation Between Votes and Rating")
plt.show()
```
