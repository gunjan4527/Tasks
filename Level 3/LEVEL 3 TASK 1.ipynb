```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load the dataset
dataset = pd.read_csv(r"C:\Users\gunjan kumar\OneDrive\Documents\Dataset .csv")
dataset.head()
```

```python
# Display basic information about the dataset

dataset.info(),dataset.head()
```

```python
# count occurrences of rating text categories

rating_counts = dataset["Rating text"].value_counts()
```

```python
# visualize the distribution of rating text categories

plt.figure(figsize=(10,5))
sns.barplot(x=rating_counts.index, y=rating_counts.values)
plt.xticks(rotation=45)
plt.xlabel("Rating Category")
plt.ylabel("Count")
plt.title("Distribution of Rating Text Categories")
plt.show()
```

```python
# Display the most common rating text categories

rating_counts
```

```python
from collections import Counter
```

```python
# Define positive and negative ratings

positive_ratings = ["Excellent", "Very Good", "Good"]
negative_ratings = ["Average", "Poor"]
```

```python
# Count occurrences of positive and negative words

positive_words = " ".join(dataset[dataset["Rating text"].isin(positive_ratings)]["Rating text"]).lower().split()
negative_words = " ".join(dataset[dataset["Rating text"].isin(negative_ratings)]["Rating text"]).lower().split()
```

```python
# Get most common words

positive_word_counts = Counter(positive_words).most_common(5)
negative_word_counts = Counter(negative_words).most_common(5)
positive_word_counts, negative_word_counts
```

```python
# calculate the length of rating text

dataset["Rating text length"] = dataset["Rating text"].apply(lambda x: len(x.split()))
```

```python
# Compute the average length of rating text

average_length = dataset["Rating text length"].mean()
```

```python
# visualize the relationship between text length and aggregate rating

plt.figure(figsize=(10,5))
sns.boxplot(x="Rating text", y="Rating text length", data=dataset, order=rating_counts.index)
plt.xticks(rotation=45)
plt.xlabel("Rating Category")
plt.ylabel("Rating Text Length")
plt.title("Relationship Between Text length and Rating Category")
plt.show()

average_length
```
