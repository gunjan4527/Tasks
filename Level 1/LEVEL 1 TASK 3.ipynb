```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# load the dataset
dataset = pd.read_csv(r"C:\Users\gunjan kumar\OneDrive\Documents\Dataset .csv")
dataset.head()
```

THE DISTRIBUTION OF PRICE RANGES AMONG THE RESTAURANTS

```python
# price range counts

price_ranges_per_counts = dataset['Price range'].value_counts().sort_index()
```

```python
# Create bar chart

colors = ['red', 'green', 'yellow', 'blue']
plt.bar(price_ranges_per_counts.index, price_ranges_per_counts.values, color=colors)
plt.title("The Distribution of Price Ranges Among The Restaurants")
plt.xlabel("Price Ranges")
plt.ylabel("Count")
plt.show()
```

CALCULATE THE PERCENTAGE OF RESTAURANTS IN EACH PRICE RANGE CATEGORY

```python
# calculate total number of restaurants

total_restaurants = len(dataset)
print("Total number of restaurants:", total_restaurants)

```

```python
# calculate the percentage of restaurants in each price range category

percentage_per_price_range = (price_ranges_per_counts / total_restaurants) * 100
print("The percentage of Restaurants in Each Price Range Category")
print(percentage_per_price_range)
```
